# Class Table

```python
engine = create_engine(f"mysql://{remote_db_user}:{remote_db_pwd}@{remote_db_endpoint}:{remote_db_port}/{remote_db_name}")
conn = engine.connect()
```

```python
class_data = pd.read_sql("SELECT * FROM class", conn)
class_data.head(30)
```

```python
class_data = class_data.rename(columns = {
    "ID_Class":"ID_Class__c",
    "ID_Course":"ID_Course__c",
    "Section":"Section__c",
    "StartDate":"Start_Date__c",
    "EndDate":"EndDate__c"
})
class_data.head()
```

```python
class_data["Start_Date__c"] = class_data["Start_Date__c"].astype(str)
class_data["EndDate__c"] = class_data["EndDate__c"].astype(str)
```

```python
query = sf.query("SELECT ID_Course__c FROM Course1__c")
query
```

```python
course_id_dict = {}
i = 0
while i < len(query['records']):
    url = query['records'][i]['attributes']['url'].split('/')[6]
    ID_Course__c = int(query['records'][i]['ID_Course__c'])
    print(url,ID_Course__c)
    course_id_dict[ID_Course__c]=url
    i = i+1
print(course_id_dict)
```

```python
class_load = class_data.to_dict('records')
class_load
```

```python
for index, row in class_data.iterrows():
    class_data['ID_Course__c'][index] = course_id_dict[row['ID_Course__c']]
class_data
```

```python
for index, row in class_data.iterrows():

    record = {
        "ID_Class__c": row["ID_Class__c"],
        "ID_Course1__c": row["ID_Course__c"],
        "Section__c": row["Section__c"],
        "StartDate__c":row["Start_Date__c"],
        "EndDate__c": row["EndDate__c"]
    }
    try:
        sf.Class__c.create(record)
    except Exception as e:
        print(e)
```
